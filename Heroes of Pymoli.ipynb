```python
Purchase_Count=purchase_data["SN"].value_counts()
Purchase_Count_df=pd.DataFrame(Purchase_Count)
Purchase_Count_df['Total Purchase Count']=purchase_data.groupby(['SN'])['Price'].sum()
Purchase_Count_df['Average Purchase Price']= round(Purchase_Count_df['Total Purchase Count']/Purchase_Count_df['SN'],2)
Purchase_Count_df
final_purchase= Purchase_Count_df.sort_values(by='Total Purchase Count', ascending=False).head(5)
final_purchase
```

```python
most_popular=purchase_data["Item ID"].value_counts()
most_popular_df=pd.DataFrame(most_popular)
most_popular_df['Total Purchase Value']=purchase_data.groupby(['Item ID'])['Price'].sum()
most_popular_df['Item Price']=purchase_data.groupby(['Item ID'])['Price'].mean()
most_popular_df.head(5)
```

```python
most_popular_final= most_popular_df.sort_values(by='Total Purchase Value', ascending=False).head(5)
most_popular_final
```
